```python
import pickle
import pandas as pd 

with open ('audio_features.pickle', 'rb') as handle:
    audio_features = pickle.load(handle)

audio_features_df = pd.DataFrame.from_dict(audio_features, orient='index')

audio_features_df.head()
```

```python
# get the total number of songs in the dataset
num_songs = len(audio_features_df)
print(num_songs)
```

```python
tags = list(playlist_cleaned_dict.keys())
audio_features_df.add(tags, axis=1)

# open the song set dictionary
with open('song_set.pickle', 'rb') as handle:
    songSet = pickle.load(handle)
```

```python
import random
# for each tag, add it as a column to the dataframe
for tag in tags:
    audio_features_df[tag] = 0

# for each song, randomly assign 1 to 5 tags to it
for index, row in audio_features_df.iterrows():
    num_tags = random.randint(1, 5)
    for i in range(num_tags):
        tag = random.choice(tags)
        audio_features_df.loc[index, tag] = 1

audio_features_df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

df = audio_features_df.copy()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[tags], df[['accousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']], test_size=0.2, random_state=42)

# Train a linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Evaluate the model on the testing set
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
print('Mean squared error:', mse)

# Use the model to predict audio feature values for new songs
# create a new dataframe with all the columns of the original dataframe except for the audio features, song name, and artist name
new_df = pd.DataFrame(columns=df.columns[9:])
# set the values of the new dataframe to 0
new_df.loc[0] = 0
# set the values of the new dataframe for the tags to 1
new_df.loc[0, 'rock'] = 1
new_df.loc[0, 'pop'] = 1
new_features = model.predict(new_df)
print('Predicted audio features:', new_features)
# given the predicted audio features, see if there;s a song in the dataset that matches those features
```

```python
# redo the above but using a neural network instead of linear regression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error

df = audio_features_df.copy()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[tags], df[['accousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']], test_size=0.2, random_state=42)

# Train a neural network model
model = MLPRegressor(hidden_layer_sizes=(100, 100, 100), max_iter=1000, random_state=42)
model.fit(X_train, y_train)

# Evaluate the model on the testing set
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

print('Mean squared error:', mse)

# Use the model to predict audio feature values for new songs
new_df = pd.DataFrame(columns=df.columns[9:])
# set the values of the new dataframe to 0
new_df.loc[0] = 0
# set the values of the new dataframe for the tags to 1
new_df.loc[0, 'rock'] = 1
new_df.loc[0, 'pop'] = 1
new_features = model.predict(new_df)
print('Predicted audio features:', new_features)
```

```python
# do the same thing but using keras
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from keras.models import Sequential
from keras.layers import Dense

df = audio_features_df.copy()

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df[tags], df[['accousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 'tempo', 'valence']], test_size=0.2, random_state=42)

# Train a neural network model
model = Sequential()
model.add(Dense(100, input_dim=len(tags), activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(9, activation='linear'))
model.compile(loss='mean_squared_error', optimizer='adam')

model.fit(X_train, y_train, epochs=1000, verbose=0)

# Evaluate the model on the testing set
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

print('Mean squared error:', mse)

# Use the model to predict audio feature values for new songs
# create a new dataframe with all the columns of the original dataframe except for the audio features, song name, and artist name
new_df = pd.DataFrame(columns=df.columns[9:])
# set the values of the new dataframe to 0
new_df.loc[0] = 0
# set the values of the new dataframe for the tags to 1
new_df.loc[0, 'rock'] = 1
new_df.loc[0, 'pop'] = 1
new_features = model.predict(new_df)
print('Predicted audio features:', new_features)
```
